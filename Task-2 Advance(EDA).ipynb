## 1. Given a valid (IPv4) IP address, return a defanged version of that IP address.
    A defanged IP address replaces every period "." with "[.]".

```python
class Solution:
    def defangIPaddr(self, address: str) -> str:
        return address.replace(".", "[.]")
address = input()
solution = Solution()
print(solution.defangIPaddr(address))
```

## 2. Given an array nums of integers, return how many of them contain an even number of digits.

```python
class Solution:
    def findNumbers(self, nums):
        count = 0
        for n in nums:
            if len(str(n)) % 2 == 0:  # check even number of digits
                count += 1
        return count
nums = list(map(int, input().split()))

solution = Solution()
print(solution.findNumbers(nums))
```

## 3. Given an array of integers nums, return the number of good pairs. 
    A pair (i, j) is called good if nums[i] == nums[j] and i < j.

```python
from collections import defaultdict

class Solution:
    def numIdenticalPairs(self, nums):
        freq = defaultdict(int)
        good_pairs = 0
        
        for num in nums:
            good_pairs += freq[num]   # count how many same numbers seen before
            freq[num] += 1
            
        return good_pairs


nums = list(map(int, input().split()))

solution = Solution()
print(solution.numIdenticalPairs(nums))
```

## 4. Given the array nums, for each nums[i] find out how many numbers in the array are smaller than it. That is, for each nums[i] you have to count the number of valid j's such that j != i and nums[j] < nums[i]. Return the answer in an array.

```python
class Solution:
    def smallerNumbersThanCurrent(self, nums):
        n = len(nums)
        ans = [0] * n
        for i in range(n):
            cnt = 0
            for j in range(n):
                if nums[j] < nums[i]:
                    cnt += 1
            ans[i] = cnt
        return ans

nums = list(map(int, input().split()))

solution = Solution()
result = solution.smallerNumbersThanCurrent(nums)

print(result)
```

## 5. Given an integer number n, return the difference between the product of its digits and the sum of its digits.

```python
class Solution:
    def subtractProductAndSum(self, n):
        product = 1
        s = 0

        while n>0:
            digit = n % 10
            product *= digit
            s += digit
            n //= 10
        return product - s

n = int(input())

solution = Solution()
print(solution.subtractProductAndSum(n))
```

## 6. You are given an integer n and an integer start. Define an array nums where nums[i] = start + 2 * i (0-indexed) and n == nums.length. Return the bitwise XOR of all elements of nums.



```python
class Solution:
    def xorOperation(self, n, start):
        xor_result = 0
        for i in range(n):
            xor_result ^= (start + 2 * i)
        return xor_result

n = int(input())
start = int(input())

solution = Solution()
print(solution.xorOperation(n, start))
```
